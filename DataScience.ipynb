6.8-Assimetria e Curtose

```python
import seaborn as sns#biblioteca

dados = pd.read_csv('/content/drive/MyDrive/DataScience/dados_saude_manipulado.csv', sep=',')#leitura do arquivo csv separado por ,

print('Idade Gestacional')

assimetria = dados.IG_OBSTETRA.skew()
print('Assimetria: ', round(assimetria, 3))

curtose = dados.IG_OBSTETRA.kurtosis()
print('Curtose: ', round(curtose, 3))

print('\nIdade da Paciente')

assimetria = dados.IDADE.skew()
print('Assimetria: ', round(assimetria, 3))

curtose = dados.IDADE.kurtosis()
print('Curtose: ', round(curtose, 3))
```
